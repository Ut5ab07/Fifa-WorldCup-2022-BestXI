Filtering players based on minutes played and filling missing data with 0


```python
filtered_midfielders = df[(df['minutes']>=360)].copy()
filtered_midfielders.fillna(0, inplace=True)
filtered_midfielders.head()
```

```python
#Applying the formula to calculate score to rank players

filtered_midfielders['score']=(
    0.18 * filtered_midfielders['chance_created_per90']+
    0.15 * filtered_midfielders['big_chance']+
    0.15 * filtered_midfielders['pass_per90']+
    0.12 * filtered_midfielders['assists_per90']+
    0.10 * filtered_midfielders['xg_assist_per90']+
    0.08 * filtered_midfielders['goals_per90']+
    0.08 * filtered_midfielders['npxg_per90']+
    0.07 * filtered_midfielders['tackles_per90']+
    0.07 * filtered_midfielders['successful_dribbles']
)
```

```python
best_midfielders = filtered_midfielders.sort_values(by='score', ascending=False).head(10)
print (best_midfielders[['player','score']])
```

```python
#importing matplotlib for visaulization
import matplotlib.pyplot as plt

plt.figure(figsize=(20,6))
plt.bar(best_midfielders['player'], best_midfielders['score'],color= 'skyblue')
plt.xlabel('Score')
plt.ylabel('Player')
plt.title('Top 10 Midfielders- FIFA World Cup 2022')
plt.show()
```
